```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
data= pd.read_csv('E:\T v\join data.csv')
```

```python
data1= pd.DataFrame(data)
```

```python
data.info()
```

```python
x=data1[['Product review available','Due to covid 19 you do online shopping','Free Delivery','EMI option available','Notification of sale','Discount'
,'Easy to buy/one click buy','Offers a Range of products','Time Saving',' What is your Family Size','What is your level of Education','What is your Age Group','Gender']]
y=data1[['If you buy from Ecommerce website how frequently you buy']]
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.15,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
#instantiate the model
log_regression = LogisticRegression()

#fit the model using the training data
model=log_regression.fit(X_train,y_train)

#use model to make predictions on test data
y_pred = log_regression.predict(X_test)
```

```python
y_pred
```

```python
model.coef_
```

```python
cnf_matrix = pd.DataFrame(metrics.confusion_matrix(y_test, y_pred),index=["Actual: 0","Actual:1","Actual:2","Actual:3"],columns=["predicted:0","predicted:1","predicted:2","predicted:3"])
cnf_matrix
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score 
```

```python
print("Mean Absolute Error:", mean_absolute_error(y_test,y_pred))
```

```python
print("Mean Squared Error:",mean_squared_error(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import GridSearchCV
```

```python
#List Hyperparameters that we want to tune.
leaf_size = list(range(1,7))
n_neighbors = list(range(1,14))
p=[1,2]
#Convert to dictionary
hyperparameters = dict(leaf_size=leaf_size, n_neighbors=n_neighbors, p=p)
#Create new KNN object
knn_2 = KNeighborsClassifier()
#Use GridSearch
clf = GridSearchCV(knn_2, hyperparameters, cv=10)
#Fit the model
best_model = clf.fit(x,y)
#Print The value of best Hyperparameters
print('Best leaf_size:', best_model.best_estimator_.get_params()['leaf_size'])
print('Best p:', best_model.best_estimator_.get_params()['p'])
print('Best n_neighbors:', best_model.best_estimator_.get_params()['n_neighbors'])
```

```python
y_predHP = knnHP.predict(X_test)
y_predHP
```
